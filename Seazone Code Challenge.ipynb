## Importando biblioteca e lendo os arquivos

## Questão 1 - Ordenar os bairros em ordem crescente de número de listings

```python
bairros_ordernados = pd.DataFrame({'Quantidade': details['suburb'].value_counts(ascending=True)})
dfi.export(bairros_ordernados, 'bairros_ordenados.png')
bairros_ordernados
```

```python
grafico = sns.barplot(x = bairros_ordernados.index, y = bairros_ordernados['Quantidade']);
grafico.set(xlabel="Bairros", ylabel="Quantidade de anúncios")
plt.xticks(rotation=10)
plt.savefig('grafico1')
plt.show()
```

## Questão 2 - Ordenar os bairros em ordem crescente de faturamento médio dos listings

```python
dados = price.merge(details, on='airbnb_listing_id')
dados.head(10)
```

```python
alugados = dados[dados['occupied'] == 1]
alugados
```

```python
faturamento_anuncios = pd.pivot_table(alugados, values = ['airbnb_listing_id','price_string','occupied'],index='suburb', aggfunc = {'price_string': np.sum, 'occupied': np.sum, 'airbnb_listing_id': pd.Series.nunique})
faturamento_anuncios['Faturamento_anuncios'] = (faturamento_anuncios['price_string'] / faturamento_anuncios['airbnb_listing_id']).round(2)
dfi.export(faturamento_anuncios[['Faturamento_anuncios']].sort_values(by = 'Faturamento_anuncios', ascending = True), 'Faturamento_anuncios.png')
faturamento_anuncios[['Faturamento_anuncios']].sort_values(by = 'Faturamento_anuncios', ascending = True)
```

```python
grafico2 = sns.barplot(x = faturamento_anuncios.index, y = faturamento_anuncios['Faturamento_anuncios'], order=faturamento_anuncios.sort_values('Faturamento_anuncios').index);
grafico2.set(xlabel="Bairros", ylabel="Faturamento Médio por Anúncio")
plt.xticks(rotation=10)
plt.savefig('grafico2')
plt.show()
```

### Centro é o bairro com o menor faturamento médio dos anúncios, e Jurerê é o com o maior

## Questão 3 - Descobrir correlações entre as características de um anúncio e seu faturamento

```python
alugados.corr()
```

### Com relação ao faturamento, já notamos anteriormente uma relação com os bairros, tendo-se ordenado os com maior faturamento, e com relação as outras características?


```python
quartos_faturamento = pd.pivot_table(alugados, values = ['price_string','airbnb_listing_id'], index = 'number_of_bedrooms', aggfunc = {'price_string': np.sum,'airbnb_listing_id': pd.Series.nunique}).reset_index()
quartos_faturamento['Faturamento_quartos'] = quartos_faturamento['price_string']/quartos_faturamento['airbnb_listing_id']
dfi.export(quartos_faturamento[['number_of_bedrooms','Faturamento_quartos']].sort_values(by = 'Faturamento_quartos', ascending = True), 'quartos_faturamento.png')
quartos_faturamento[['number_of_bedrooms','Faturamento_quartos']].sort_values(by = 'Faturamento_quartos', ascending = True)
```

```python
quartos_faturamento[['number_of_bedrooms','Faturamento_quartos']].corr()
```

```python
ax = sns.lmplot(data=quartos_faturamento, x='number_of_bedrooms', y='Faturamento_quartos')
ax.fig.set_size_inches(12,6)
ax.fig.suptitle('Reta de Regressão - Quartos x Faturamento Médio', fontsize=12, y=1.02)
ax.set_xlabels('Número de Quartos', fontsize=14)
ax.set_ylabels('Faturamento Médio', fontsize=14)
plt.savefig('grafico3')
ax
```

```python
banheiros_faturamento = pd.pivot_table(alugados, values = ['price_string','airbnb_listing_id'], index = 'number_of_bathrooms', aggfunc = {'price_string': np.sum,'airbnb_listing_id': pd.Series.nunique}).reset_index()
banheiros_faturamento['Faturamento_banheiros'] = banheiros_faturamento['price_string']/banheiros_faturamento['airbnb_listing_id']
dfi.export(banheiros_faturamento[['number_of_bathrooms','Faturamento_banheiros']].sort_values(by = 'Faturamento_banheiros', ascending = True), 'banheiro_faturamento.png')
banheiros_faturamento[['number_of_bathrooms','Faturamento_banheiros']].sort_values(by = 'Faturamento_banheiros', ascending = True)
```

```python
banheiros_faturamento[['number_of_bathrooms','Faturamento_banheiros']].corr()
```

```python
ax = sns.lmplot(data=banheiros_faturamento, x='number_of_bathrooms', y='Faturamento_banheiros')
ax.fig.set_size_inches(12,6)
ax.fig.suptitle('Reta de Regressão - Banheiros x Faturamento Médio', fontsize=12, y=1.02)
ax.set_xlabels('Número de Banheiros', fontsize=14)
ax.set_ylabels('Faturamento Médio', fontsize=14)
plt.savefig('grafico4')
ax
```

### Existe uma correlação positiva entre o número de quartos e o número de banheiros com o faturamento

## Questão 4 - Antecedência média das reservas, e se é maior ou menor para finais de semana

```python
alugados.head()
```

### booked_on e date estão em formatos diferentes, é necessário converter para o formato de data.

```python
alugados['date'] = pd.to_datetime(alugados['date'])
alugados['booked_on'] = pd.to_datetime(alugados['booked_on'])
alugados['dia_semana'] = alugados['date'].dt.dayofweek
alugados['antecedencia'] = alugados['date']-alugados['booked_on']
print('Antecedência média:',alugados['antecedencia'].mean())
```

### A antecedência média das reservas é em torno de 37 dias e 20 horas

```python
alugados[['date', 'booked_on', 'antecedencia', 'dia_semana']]
```

```python
alugados_fim_de_semana = alugados.loc[alugados['dia_semana'] > 4]
alugados_semana = alugados.loc[alugados['dia_semana'] < 5 ]
```

```python
antecedencia_fim_de_semana = alugados_fim_de_semana['antecedencia'].mean()
antecedencia_semana = alugados_semana['antecedencia'].mean()
antecedencia_semana/antecedencia_fim_de_semana
```

```python
print('Diferença entre antecedência para semana e finais de semana:',antecedencia_semana - antecedencia_fim_de_semana)
```

```python
print('Antecedência média para dias de semana:',antecedencia_semana)
```

```python
print('Antecedência média para finais de semana:',antecedencia_fim_de_semana)
```

### Para os finais de semana, a antecedência média de reservas é menor em torno de 1 dia.
